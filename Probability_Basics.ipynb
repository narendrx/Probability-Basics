# Probability Basics


1. A die is rolled. What is the probability of getting:
(a) An even number
(b) A number greater than 4
- Total outcomes S = \{1, 2, 3, 4, 5, 6\}, so n(S) = 6.
(a) Even number: E = \{2, 4, 6\}. P(\text{Even}) = \frac{3}{6} = \mathbf{0.5} (or 50%).
(b) Greater than 4: E = \{5, 6\}. P(> 4) = \frac{2}{6} = \mathbf{\frac{1}{3}} (or approx 33.3%).


2. In a class of 50 students:
20 like Mathematics (M)
15 like Science (s)
5 like both subjects
What is the probability that a student chosen at random likes Mathematics or Science?
- Total students (N) = 50.
P(M) = \frac{20}{50}, P(S) = \frac{15}{50}, P(M \cap S) = \frac{5}{50}.
Formula: P(M \cup S) = P(M) + P(S) - P(M \cap S)
Calculation: \frac{20}{50} + \frac{15}{50} - \frac{5}{50} = \frac{30}{50} = \mathbf{0.6} (or 60%).

3. A bag has 3 red and 2 blue balls. If one ball is drawn randomly and is red, what is the probability that the next ball is also red (without replacement)?
- Initial: 3 Red, 2 Blue (Total 5).
One Red is drawn and not replaced.
Remaining: 2 Red, 2 Blue (Total 4).
Probability the next is Red: \frac{2}{4} = \mathbf{0.5} (or 50%).

4. The population of a school is divided into 60% boys and 40% girls. If you want equal representation of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling? Why?
- Method: Stratified Sampling.
Why: Because the population has distinct subgroups (60% boys, 40% girls). Stratified sampling ensures the sample reflects these specific proportions, whereas Simple Random Sampling might accidentally pick too many of one group in a small sample.

5. The average height of 1000 students = 160 cm. A sample of 100 students shows an average height = 158 cm.
Find the sampling error.
- \text{Sampling Error} = |\text{Population Mean} - \text{Sample Mean}|
- Calculation: |160\text{ cm} - 158\text{ cm}| = \mathbf{2\text{ cm}}.

6. The population mean salary is 250,000 with o = 25,000. If we take a sample of 100 employees, what is the standard error of the mean (SEM)?
- SEM = \frac{\sigma}{\sqrt{n}}
\sigma = 5,000, n = 100.
SEM = \frac{5000}{\sqrt{100}} = \frac{5000}{10} = \mathbf{₹500}.

7. In a group of 100 students:
40 like Cricket (C)
30 like Football (F)
10 like both Cricket and Football
Find the probability that a student likes at least one sport.
- Total students = 100.
- n(C \cup F) = n(C) + n(F) - n(C \cap F)
- 40 + 30 - 10 = 60.
- Probability: \frac{60}{100} = \mathbf{0.6} (or 60%)

8. From a deck of 52 cards, two cards are drawn without replacement. What is the probability that both are Aces?
- P(1st Ace) = \frac{4}{52}
P(2nd Ace | 1st Ace) = \frac{3}{51}
Total Probability: \frac{4}{52} \times \frac{3}{51} = \frac{12}{2652} = \mathbf{\frac{1}{221}} (approx 0.0045).

9. A factory produces bulbs with 2% defective rate. If 5 bulbs are chosen at random, what is the probability that all are non-defective?
- Defective rate = 2% (0.02). Non-defective rate (p) = 98\% (0.98).
For 5 bulbs to all be non-defective: 0.98^5
Calculation: \mathbf{\approx 0.9039} (or 90.39%).
